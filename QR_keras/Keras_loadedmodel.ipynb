# load trained model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#keras model es
model_loaded_es = keras.models.load_model("keras_model_es.keras")
```

```python
#load test dataset
df_test = pd.read_csv("../../digit-recognizer/test.csv")
df_test.head()
```

```python
#extract features and normalise
x_test = df_test.values / 255.0 # normalise to [0,1]

# reshape to 28x28x1 for CNN
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
predictions = model_loaded_es.predict(x_test)
```

```python
predictions_values = predictions.argmax(axis = 1)
```

```python
loadedmodel_es = pd.DataFrame({
    "ImageId": range(1, len(predictions)+1),  # 1-based indexing
    "Label": predictions_values
})

loadedmodel_es.to_csv("QR_keras_loadedmodel_es.csv", index=False)
```
